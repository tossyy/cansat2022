```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from math import *

with open('/Users/k.sakamoto/Desktop/log/target_posision.txt', mode='r') as f:
    lines = [s.strip() for s in f.readlines()]
    target = (float(lines[0]), float(lines[1]))


df = pd.read_csv('/Users/k.sakamoto/Desktop/log/phase6.csv')

plt.figure(figsize=(12, 8))
plt.plot(df['longitude'], df['latitude'], marker='o')
plt.plot(target[1], target[0], marker='^', markersize=10)
for index, row in df.iterrows():
    plt.arrow(row[2], row[1], row[4]/3000000*np.cos(row[3]), row[4]/3000000*np.sin(row[3]), width=0.0000005)

plt.xlabel('longitude')
plt.ylabel('latitude')
```
